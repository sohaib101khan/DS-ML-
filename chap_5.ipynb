```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


fish_data=pd.read_csv('/content/Fish_Sample_Data.csv')
fish_data.head()

```

##SCATTERPLOT FOR SEABORN
in this it show the relation between the length of the fish and its weeight in this each dot represent a fish when you move to right side where the fish are longer and dots also go higher its mean they are heavier

its mean the longer the fish is the higher the weight will be This means there's a positive relationship between the fish's length and weight: more length, more weight.

###BELOW CODE
in this code first we take seaborn set darkgrid it gives us the grey background and palette give us soft colors and al this bar is to find out the wight of the fish and how much the longer fish is

```python
sns.set(style="darkgrid")
plot = sns.scatterplot(
    data=fish_data,
    x='Length_A',
    y='Mass_g',
    hue='Type',
    palette='pastel',   # softer colors
    s=70,
    edgecolor='black'
)
plot.set_title('Analysis of Fish Length vs. Weight', fontsize=14, fontweight='bold')
plot.set_xlabel('Fish Length (cm)', fontsize=12)
plot.set_ylabel('Fish Weight (g)', fontsize=12)
```

###BELOW CODE
in this we are distrubuting the mass of fish type and we use loop to show the data of all types of fishes

```python


plt.figure(figsize=(10, 6))
ax = sns.boxplot(data=fish_data, x='Type', y='Mass_g',palette='pastel')
medians = fish_data.groupby('Type')['Mass_g'].median().values
for i, median in enumerate(medians):
    ax.text(i, median + 10, f'Median = {median:.1f}',
            horizontalalignment='center', color='grey', weight='bold',fontsize=8)
plt.title('Mass Distribution by Fish Type')
plt.xlabel('Fish type')
plt.ylabel('Mass (g)')

plt.show()

```

##HISTOGRAM
in this we will compare the weight of two type of fishes snapper and salmonn the purple bar represent snapper and green bar represent salmon

Salmon : Mostly light (0–2), with a small spread in values.
Snapper : road range (1–4), includes some higher values, moderate variation.

###BELOW CODE
in this we make comparison in this through histogram bar

```python
fish_snapper = fish_data.loc[fish_data['Type'] == 'Snapper']
fish_salmon = fish_data.loc[fish_data['Type'] == 'Salmon']

ax = sns.histplot(data=fish_salmon, x='Mass_g', color='red', alpha=0.5, label='Salmon')
ax = sns.histplot(data=fish_snapper, x='Mass_g', color='blue', alpha=0.5, label='Snapper')

ax.set_xlabel('Fish mass (grams)', fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.set_title('Comparison of Fish mass for Salmon and Snapper', fontsize=20)

plt.legend()
plt.show()

```

###BELOW CODE
in the code we make loop so it can give us the mass distribution data of all fish types

```python



Type_list = fish_data['Type'].unique()

fig, axes = plt.subplots(nrows=len(Type_list), figsize=(8, 4 * len(Type_list)))

for i, Type in enumerate(Type_list):
    subset = fish_data[fish_data['Type'] == Type]['Mass_g']
    axes[i].hist(subset, bins=10, color='lightgreen', edgecolor='black')
    axes[i].set_title(f'{Type} - Mass Distribution')
    axes[i].set_xlabel('Mass (g)')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()



```

##correlation heatmap in seaborn

This heatmap shows how different numbers in our fish data are connected. In our data, weight and length are strongly linked longer fish are usually heavier.
All size measurements are also linked with each other


###BELOW CODE
we can see every different  number has different color becasue if heatmap

```python
numeric_data = fish_data.select_dtypes(include='number')

corr = numeric_data.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Correlation Heatmap', fontsize=14, pad=12)
plt.tight_layout()
plt.show()
```
