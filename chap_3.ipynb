##STACKED BAR CHAT

we use stacked bar When you want to compare groups e.g., professions And see how sub-categories e.g., scores are distributed within those groups like in our case we will use it for to see the score distributed in different profession

###IN BLEOW CODE
as we can see the first line it Create a cross-tab table between Profession and Presentation Score

stacked=True = stacks the different score levels on top of each other in each bar.

colormap='Pastel2' => gives soft and nice colors to the stacks for better visibility.

```python

filtered_df = df[(df['Presentation_Score'] >= 70) & (df['Presentation_Score'] <= 100)]

profession_counts = filtered_df['Profession'].value_counts()

profession_counts.plot(kind='bar', figsize=(8, 5), color='skyblue',)
plt.title('Number of Participants by Profession (Score 70 to 100)')
plt.xlabel('Profession')
plt.ylabel('Number of Participants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Z-SCORE
z-score tells us how far the value is from mean(average) If Z-score is more than +2 or less than -2, we call it an outlier we use Z-score when To find unusual values (like cheating, errors, very poor scores)
Helps in cleaning and understanding your data

###CODE LOGIC
in first line we takes the absolute value of the Z-score.So both very high and very low scores are treated equally. >2 means:
Keep only those rows where the score is more than 2 standard deviations away from the mean

```python
df['Score_Zscore'] = zscore(df['Presentation_Score'])

df[df['Score_Zscore'].abs() > 2]

```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
sns.scatterplot(data=df, x='Age', y='Presentation_Score', hue='Profession', ax=axes[0])
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Presentation Score')
axes[0].set_title('Age vs Presentation Score by Profession')

# Bar plot
sns.barplot(data=df, x='Profession', y='Feedback_Rating', ax=axes[1], palette='Set2',estimator='mean', errorbar=None)
axes[1].set_xlabel('Profession')
axes[1].set_ylabel('Average Feedback Rating')
axes[1].set_title('Feedback Rating by Profession')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

###BELOW CODE
in below code we are checking the per professionn score so we use a loop for this to tae a data from all lists

```python
professions = df['Profession'].unique()

fig, axes = plt.subplots(3, 2, figsize=(10, 9))
axes = axes.flatten()

for i, profession in enumerate(professions):
    df_prof = df[df['Profession'] == profession].copy()
    df_prof['Participant_ID'] = range(1, len(df_prof) + 1)
    sns.lineplot(
        data=df_prof,
        x='Participant_ID',
        y='Presentation_Score',
        color=sns.color_palette("Set2")[i],
        linewidth=2,
        ax=axes[i]
    )
    axes[i].set_title(f'{profession} Presentation Scores', fontsize=12)
    axes[i].set_xlabel('Participant')
    axes[i].set_ylabel('Presentation Score')

for j in range(len(professions), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

##Violin plot
 violin plot is like a box plot and density plot combined.

The distribution of the data (like a histogram, but smoothed with a kernel density estimate).

The median and quartiles (like a box plot).

The shape of the data — wider parts mean more data points there.

we use this to see spread and distrubution  shape of numeric data

```python
sns.violinplot(x='Profession', y='Feedback_Rating', data=df, palette='pastel')
plt.title('Feedback Rating Distribution by Profession')
plt.xlabel('Profession')
plt.ylabel('Feedback Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
