##CHANGING DATA FORMAT
in this we are converting the float into integer and we fill the missing values with fillna(0) and change its dtype into integer with astype(int)

###BELOW CODE LOGIC
in first we add a csv file and we print a before message and then we check the dtype of both Age and Feedback_Rating and in next line we add the value 0 in the null rows and in next step we convert it into integer same we do for Age and in last we again change dtype and check the values we fill and make it int


```python
df_33 = pd.read_csv('/content/techfest_participants_300(2).csv')
print("before")
print(df_33['Feedback_Rating'].dtype)
print(df_33['Age'].dtype)

df_33['Feedback_Rating'] = df_33['Feedback_Rating'].fillna(0)
df_33['Feedback_Rating'] = df_33['Feedback_Rating'].astype(int)

df_33['Age'] = df_33['Age'].fillna(0)
df_33['Age'] = df_33['Age'].astype(int)


print("\nafter")
print(df_33['Feedback_Rating'].dtype)
print(df_33['Feedback_Rating'].head())

print(df_33['Age'].dtype)
print(df_33['Age'].head())
```

##GROUPING AND AGGREGATION
###MEAN AGE BY CITY
this will group all the rows in the dataset by the city column like it will make group for people of every city and then ake the average age of those participants



###BELOW CODE
in this first we take a group of participants from each city and  then we take the average agee of people from the city by using mean and after this we use reset_index() to convert the data back into its actual dataframe

###SCORE BY PROFESSION

##CAEGORIZE SCORE INTO BINS
in this we are dividing the score in three categories

low from 0 to 65

medium from 65 to 80

high from 80 to 100

bins [0,65,80,100]

labels[low,medium,high]

df=['score_lvl]
creates a new column in the DataFrame called Score_Level, which will store:

Low / Mid / High based on each person’s score.

###BELOW CODE
forst we make a new column in our dataframe name as score_lvl
and used pd.cut(this is a pandas function that splits numeric values into categories (bins)) and then we make a bin and give it labels

```python
df['Score_Lvl'] = pd.cut(df['Presentation_Score'],
                           bins=[0, 70, 85, 100],
                           labels=['Low', 'Mid', 'High'])
df[['Presentation_Score', 'Score_Lvl']].head()
```

```python
df.head()
```

##Seaborn Count Plot — a type of bar chart
in this it will show hoe many participants will lie in each score_lvl so for this reason we are now making a bar to take a complete analysis about this

###CODE LOGIC
the first line sns...........grid" it gives the plot a white background with light grid lines and in last

And in loop the first line Go through each bar in the current chart one by one.and in 2nd line , line gets the height of the bar. and the 3rd line  This line finds the middle x-position of the bar. and the last line in the loop This line actually adds the number on the bar.




```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

sns.countplot(
    data=df,
    x='Score_Lvl',
    hue='Score_Lvl',
    palette='pastel',
    legend=False
)

plt.title("Score Lvl Distribution")
plt.xlabel("Score Lvl")
plt.ylabel("Participants")

for bar in plt.gca().patches:
    height = bar.get_height()
    x = bar.get_x() + bar.get_width() / 2
    y = height
    plt.text(x, y + 5, str(int(height)), ha='center', fontsize=11, color='black')

plt.tight_layout()
plt.show()
```
